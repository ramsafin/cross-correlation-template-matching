```matlab
addpath('../');
```

```matlab
% load an image (shelf)
I = rgb2gray(imread('../pictures/shelf.png'));
imshow(I);
```

```matlab
% load the template to search for
 t = rgb2gray(imread('../pictures/template.jpg'));
 imshow(t);
```

```matlab
% calculate the cross-correlation
c = xcorr2(t, I);

% display the surface
surf(c), shading flat;
```

```matlab
% find indices of the place where the template matches
[x_index, y_index] = find_favorite(t, I)
```

```matlab
% draw a rectangle where the template is supposed to be located
imshow(I), imrect(gca, [x_index, y_index, size(t, 2), size(t, 1)]);
```

```matlab
% subtract the mean from the template and image
t = t - mean2(t);
I = I - mean2(I);
```

```matlab
% calculate the cross-correlation
c = xcorr2(t, I);

% display the surface
surf(c), shading flat;
```

```matlab
% calculate the location
[x_index, y_index] = find_favorite(t, I)

% draw a rectangle where the template is supposed to be located
imshow(I), imrect(gca, [x_index, y_index, size(t, 2), size(t, 1)]);
```

```matlab
% load a dark shelf image
I = rgb2gray(imread('../pictures/dark shelf.jpg'));
imshow(I);
```

```matlab
% calculate the cross-correlation (normalized)
c = normxcorr2(t, I);

% display the surface
surf(c), shading flat;
```

```matlab
% find the peak
[y_peak, x_peak] = find(c == max(c(:)))

% and get indices
x_index = x_peak - size(t, 2) + 1
y_index = y_peak - size(t, 1) + 1
```

```matlab
% draw a rectangle
imshow(I), imrect(gca, [x_index, y_index, size(t, 2), size(t, 1)]);
```
