```matlab
% get a real image
I = rgb2gray(imread('../pictures/cat.jpg'));
imshow(I);
```

```matlab
% generate a gaussian 3x3 kernel
gauss_kernel = fspecial('gaussian', 3, 1)
```

```matlab
% add some noise
J = imnoise(I, 'speckle', 0.008);
imshow(J);
```

```matlab
% apply gaussian filter (convolution)
J = conv_nested(J, gauss_kernel);
imshow(J);
```
